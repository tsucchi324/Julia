```julia
function SentinelLinearSerach(A::Vector, v)
    A = [A; v]
    k = 1
    while A[k] != v
        k += 1
    end
    if k < length(A) 
        return "見つかる"
    else
        return "見つからない"
    end
end

A = rand(1:500, 200)
sort(unique(A))
```

```julia
SentinelLinearSerach(A, 496)
```

```julia
using Dates
```

```julia
format = Dates.DateFormat("m/d/Y")
```

```julia
parse(Date, date_strings[1], format)
```

```julia
parse(Date, date_strings[1], format) + Year(2000)
```

```julia
dates = parse.(Date, date_strings, format) .+ Year(2000)
```

```julia
plot(dates, US_data, xticks=dates[1:5:end], xrotation=45, leg=:topleft, 
    label="US data", m=:o)

xlabel!("date")
ylabel!("confirmed cases in US")
title!("US confirmed COVID-19 cases")
```

```julia
plot(dates, Japan_data, xticks=dates[1:5:end], xrotation=45, leg=:topleft, 
    label="Japan data", m=:o)

xlabel!("date")
ylabel!("confirmed cases in Japan")
title!("Japan confirmed COVID-19 cases")
```
